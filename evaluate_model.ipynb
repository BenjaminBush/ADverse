```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
file = open('history', 'rb')
data = pickle.load(file)
```

```python
plt.plot(data['acc'], label='Train')
plt.plot(data['val_acc'], label='Validation')
plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
```

```python
plt.plot(data['loss'], label='Train')
plt.plot(data['val_loss'], label='Validation')
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
```
